```python
import  pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("miles.csv")
df.head()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df.miles)
```

```python
sns.histplot(df.miles, kde=True)
```

```python
SAMPLE_SIZE = 50
sample_means = []
sample = df.sample(SAMPLE_SIZE, random_state=100)
sample.head()
```

```python
mean_sample = sample.miles.mean()
std_dev_sample = sample.miles.std()

mean_sample, std_dev_sample
```

```python
mean_population = df.miles.mean()
std_dev_population = df.miles.std()
mean_population, std_dev_population
```

```python
import scipy.stats as st

z_score_95_confidence = st.norm.ppf(0.975)
z_score_95_confidence
```

```python
margin_of_error = z_score_95_confidence * (std_dev_sample / np.sqrt(SAMPLE_SIZE))
margin_of_error
```

```python
upper_bound = mean_sample + margin_of_error
lower_bound = mean_sample - margin_of_error

print(f"Confidence Interval: [{lower_bound}, {upper_bound}]")
```
